```python
def map_incident_2_gmap_JS_object_str(incident_record):
    geopoint = {'lat': incident_record[0], 'lon': incident_record[1]}
    return "new google.maps.LatLng({lat}, {lon}),".format(**geopoint)

def all_geopoints_as_JS_objects_array_string(list_of_lat_lon_pairs):
    gmap_JS_object_str_df = pd.DataFrame(list_of_lat_lon_pairs).apply(map_incident_2_gmap_JS_object_str, axis=1)
    gmap_JS_object_str_list = list(gmap_JS_object_str_df)
    return str(gmap_JS_object_str_list).replace("'","").replace(",,",",")

def crime_heatmap(mapcanvas_div_id, epicenter_latitude, epicenter_longitude, geopoints_array_str):
    context = {'id': mapcanvas_div_id,
               'epicenterLat': epicenter_latitude,
               'epicenterLon': epicenter_longitude,
               'geopoints': geopoints_array_str}

    return Javascript("""
                var mapOptions = {{
                    zoom: 9,
                    center: new google.maps.LatLng({epicenterLat}, {epicenterLon})
                }};

                var carTheftMap =
                    new google.maps.Map(document.getElementById('{id}'), mapOptions);
        
                var carTheftGeopoints = {geopoints};
    
                var heatmap = new google.maps.visualization.HeatmapLayer({{
                    data: new google.maps.MVCArray(carTheftGeopoints)
                }});
    
                heatmap.setMap(carTheftMap);
            """.format(**context))
```

```python
top_communities = top_communities_with_car_theft_frequencies.keys()
which_have_car_thefts = df["community"].isin(top_communities) & contains_car_thefts
location_of_incidents = df[["lat", "lon"]][which_have_car_thefts]
    
epicenter_latitude = location_of_incidents["lat"].mean()
epicenter_longitude = location_of_incidents["lon"].mean()
geopoints_array_str = all_geopoints_as_JS_objects_array_string(location_of_incidents)

crime_heatmap("car-thefts-in-top-communities", epicenter_latitude, epicenter_longitude, geopoints_array_str)
```

## 2. Communities Ranked by the # DUIs between 10:00 PM and 2:00 AM

```python
# First, create the filters
# http://stackoverflow.com/questions/14539992/pandas-drop-rows-outside-of-time-range
indexed_by_time = df.copy().set_index(pd.to_datetime(df.time))
outside_2_to_10 = (indexed_by_time.index.indexer_between_time(time(2), time(22),
                                                              include_start=False, include_end=False))
between_2_to_10 = ~df.index.isin(outside_2_to_10)
contains_duis = df["type"].isin(["DUI"])
dui_incidents = df[df.columns][contains_duis & between_2_to_10]
```

```python
# Finally, count the frequencies of the data records grouped by community
TOP = 20
by_communities_frequencies = dui_incidents.groupby(["community"]).size()
top_communities_with_duis_frequencies = Counter(dict(by_communities_frequencies)).most_common(TOP+1)
```

Again, the community with the most incidents of the crime type is "None".<br />
Let's consider only the DUI incidents with named communities.

```python
top_communities_with_duis_frequencies_wo_none = dict(top_communities_with_duis_frequencies)
top_communities_with_duis_frequencies_wo_none.pop("NONE", None)
pd.Series(top_communities_with_duis_frequencies_wo_none).plot(kind="bar")
```
