```python
def map_incident_2_gmap_JS_object_str(incident_record):
    geopoint = {'lat': incident_record[0], 'lon': incident_record[1]}
    return "new google.maps.LatLng({lat}, {lon}),".format(**geopoint)

def all_geopoints_as_JS_objects_array_string(list_of_lat_lon_pairs):
    gmap_JS_object_str_df = pd.DataFrame(list_of_lat_lon_pairs).apply(map_incident_2_gmap_JS_object_str, axis=1)
    gmap_JS_object_str_list = list(gmap_JS_object_str_df)
    return str(gmap_JS_object_str_list).replace("'","").replace(",,",",")

def crime_heatmap(mapcanvas_div_id, epicenter_latitude, epicenter_longitude, geopoints_array_str):
    context = {'id': mapcanvas_div_id,
               'epicenterLat': epicenter_latitude,
               'epicenterLon': epicenter_longitude,
               'geopoints': geopoints_array_str}

    return Javascript("""
                var mapOptions = {{
                    zoom: 10,
                    center: new google.maps.LatLng({epicenterLat}, {epicenterLon})
                }};

                var carTheftMap =
                    new google.maps.Map(document.getElementById('{id}'), mapOptions);
        
                var carTheftGeopoints = {geopoints};
    
                var heatmap = new google.maps.visualization.HeatmapLayer({{
                    data: new google.maps.MVCArray(carTheftGeopoints)
                }});
    
                heatmap.setMap(carTheftMap);
            """.format(**context))
```

## 2. Communities Ranked by the # DUIs between 10:00 PM and 2:00 AM

```python
# First, create the filters
# http://stackoverflow.com/questions/14539992/pandas-drop-rows-outside-of-time-range
indexed_by_time = df.copy().set_index(pd.to_datetime(df.time))
outside_2_to_10 = (indexed_by_time.index.indexer_between_time(time(2), time(22),
                                                              include_start=False, include_end=False))
between_2_to_10 = ~df.index.isin(outside_2_to_10)
contains_duis = df["type"].isin(["DUI"])
dui_incidents = df[contains_duis & between_2_to_10]
```

```python
# Finally, count the frequencies of the data records grouped by community
TOP = 10
by_communities_frequencies = dui_incidents.groupby(["community"]).size()
by_communities_frequencies.sort(ascending=False)
top_communities_with_duis_frequencies = by_communities_frequencies
```

Again, the community with the most incidents of the crime type is "None".<br />
Let's consider only the DUI incidents with named communities.

```python
top_communities_with_duis_frequencies_wo_none = top_communities_with_duis_frequencies[1:TOP+1]
axis = pd.Series(top_communities_with_duis_frequencies_wo_none).plot(kind="bar")
plt.gcf().set_size_inches((9, 9))
axis.set_title("Top %d Communities with Car Thefts" % TOP)
axis.set_xlabel("Community")
axis.set_ylabel("Frequency")
```

## 3. Rhythm Maps

```python
def plot_a_heatmap(dataframe, axis, xlabel, ylabel, title):
    heatmap = axis.pcolor(dataframe, cmap=plt.cm.OrRd)
    # set titles
    axis.set_title(title,y=1.08)
    # turn off the frame, which has a gray background to separate the plots
    axis.set_frame_on(False)
    # center the major ticks at the middle of each cell
    axis.set_yticks(np.arange(dataframe.shape[0])+0.5, minor=False)
    axis.set_xticks(np.arange(dataframe.shape[1])+0.5, minor=False)
    # move x axis labels to the top, and pull it closer to the table
    axis.xaxis.tick_top()
    axis.invert_yaxis()
    # set the labels
    axis.set_xticklabels(xlabel, minor=False);
    axis.set_yticklabels(ylabel, minor=False);
    # turn off ticks
    axis = plt.gca()
    for t in axis.xaxis.get_major_ticks(): 
        t.tick1On = False 
        t.tick2On = False 
    for t in axis.yaxis.get_major_ticks(): 
        t.tick1On = False 
        t.tick2On = False
    
    return heatmap
```

```python
dayofweek_and_hour_bins_vs_counts = df[["dow", "hour"]].groupby(["dow", "hour"]).size()
dayofweek_vs_hour_bins = dayofweek_and_hour_bins_vs_counts.unstack(level=1).fillna(0)
dayofweek_vs_hour_bins.index = day_abbr
```

```python
fig, ax = plt.subplots(figsize=(12,6))
hm = plot_a_heatmap(dayofweek_vs_hour_bins, ax, xlabel=range(0,24), ylabel=day_abbr, title="Car Theft Intensity by Day of Week and Hour of Day")
cbar = plt.colorbar(hm, orientation="horizontal")
cbar.set_label("# of Car Thefts")
```

## 4. Community with Largest Change (aggregated by week of year) in # of DUIS per week

```python
all_dui_incidents = df[df.columns][contains_duis]

deltas_of_community_incidents_each_week = all_dui_incidents.groupby(["community", "week"]).size().diff() # need to drop NaN values
deltas_of_community_incidents_each_week = deltas_of_community_incidents_each_week[np.isfinite(deltas_of_community_incidents_each_week.values)]
nonnegative_deltas_of_community_incidents_each_week = deltas_of_community_incidents_each_week.abs()

change_in_communities_duis_per_week = pd.DataFrame(nonnegative_deltas_of_community_incidents_each_week).mean(level=0)
change_in_communities_duis_per_week.columns = [u"Δduis/week"]
change_in_communities_duis_per_week.sort(u"Δduis/week", ascending=False)
```
